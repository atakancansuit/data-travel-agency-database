### DeÄŸerleri okuma


Ã–nceki `get_key_value_from_city` fonksiyonunuzu kullanarak iki argÃ¼man alan yeni bir fonksiyon oluÅŸturun: `city_name` ve `key`.

EÄŸer ÅŸehir listede yoksa, fonksiyon `"*city* doesn't exist or is not covered"` dÃ¶ndÃ¼rmelidir.

Fonksiyonunuzu Ã§eÅŸitli Ã¶rneklerle test edin.

```python
def get_key_value_from_cities(city_name, key):
    for city in cities:
        if city["name"] == city_name:
            return get_key_value_from_city(city, key)
    return print(f"{city_name} doesn't exist or is not covered")
get_key_value_from_cities("Broklyn", "total_population")
```

### DeÄŸer gÃ¼ncelleme


Bazen ÅŸehir bilgileri yanlÄ±ÅŸ olabilir veya gÃ¼ncellemeye ihtiyaÃ§ duyabilir.

`modify_city_key_value` fonksiyonunu oluÅŸturun; bu fonksiyon mevcut bir anahtarÄ±n deÄŸerini yeni bir deÄŸerle deÄŸiÅŸtirir. \
Fonksiyon Ã¼Ã§ argÃ¼man alÄ±r: `city_name`, `key` ve anahtara atanacak yeni `value`.

Fonksiyon aÅŸaÄŸÄ±daki durumlardan birini belirten bir metin mesajÄ± Ã¼retecektir:

- EÄŸer ÅŸehir yoksa veya kapsanmÄ±yorsa: `"*city* doesn't exist or is not covered."`
- EÄŸer verilen anahtar yoksa: `"*key* doesn't exist."`
- EÄŸer anahtar baÅŸarÄ±yla gÃ¼ncellendiyse: `"*key* has been updated for *city*."`

```python
def modify_city_key_value(city_name, key, value):
    for city in cities:
        if city["name"] == city_name:
            if key in city:
                city[key] = value
                return f"{key} has been updated for {city_name}"
            else: 
                return f"{key} does not exist"           
    return f"{city_name} doesn't exist or is not covered"
modify_city_key_value("Brooklyn", "total_population", 3000000)
            
    
```

### Arama motorunu tasarlayÄ±n


`get_cities_nearby` adlÄ± bir fonksiyon oluÅŸturun; bu fonksiyon iki argÃ¼man alÄ±r:
- `customer_location`: [lat, long] formatÄ±nda koordinatlardan oluÅŸan bir liste


- `radius`: kilometre cinsinden eÅŸik deÄŸeri

Ve verilen `customer_location` yakÄ±nÄ±nda bulunan `ÅŸehirlerin bir listesini` dÃ¶ndÃ¼rÃ¼r.

EÄŸer hiÃ§bir ÅŸehir bulunamazsa, fonksiyon boÅŸ bir liste dÃ¶ndÃ¼rmelidir.

Test etmek iÃ§in Ã‡in'den Guangyan ÅŸehrini ([32.4355, 105.8436]) ve 500 km'lik bir yarÄ±Ã§apÄ± kullanabilirsiniz.

Xi'an ve Chengdu'nun Ã§ok uzak olmadÄ±ÄŸÄ±nÄ± gÃ¶rmelisiniz...

```python
def get_cities_nearby(customer_location,radius):
    nearby_cities = []
    for city in cities:
        city_location = city["location"]
        dist = distance.distance(customer_location, city_location).km
        if dist <= radius:
            nearby_cities.append(city["name"])
    return nearby_cities
         
```
